## Data Preprocessing

####  suppression les produits qui ont une valeur d'exportation de 0 pour toutes les années de 2014 à 2022

```python
# Sélectionner les colonnes des années de 2014 à 2022
year_columns = [f'Exported value in {year}' for year in range(2014, 2023)]

# Créer un masque pour les lignes où toutes les valeurs des colonnes des années sont 0
mask = (df_reduced[year_columns] == 0).all(axis=1)

# Filtrer les lignes à garder
df_filtered = df_reduced[~mask]

# Afficher les dimensions du DataFrame filtré
print("Dimensions du DataFrame après suppression des produits avec valeurs d'exportation de 0 pour toutes les années :", df_filtered.shape)

# Afficher les premières lignes du DataFrame filtré
print(df_filtered.head())
```

## Model training

#### Préparation des données

#### Séparer les données en ensembles d'entraînement et de test

```python

# Initialiser un dictionnaire pour stocker les données par produit
product_store = {}
# Initialiser les dataframes pour l'ensemble d'entraînement et de test
train_data = pd.DataFrame()
test_data = pd.DataFrame()

cutoff_year = 2020

# Séparer les données pour chaque produit
for product in time_series22['Product'].unique():
    # Extraire les données pour le produit courant
    product_data = time_series22[time_series22['Product'] == product]
    # Séparer les données en entraînement et test selon la date de coupe
    product_train = product_data[product_data['Year'] < cutoff_year]
    product_test = product_data[product_data['Year'] >= cutoff_year]
    # Ajouter les données au dataframe d'ensemble d'entraînement et de test
    train_data = pd.concat([train_data, product_train])
    test_data = pd.concat([test_data, product_test])
    product_store[product] = {'train': train_data, 'test': test_data}


# Réinitialiser les index
train_data = train_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)

# Afficher les dimensions des ensembles d'entraînement et de test
print("Dimensions de l'ensemble d'entraînement :", train_data.shape)
print("Dimensions de l'ensemble de test :", test_data.shape)

```

```python
# Afficher les premières lignes de l'ensemble d'entraînement
print(train_data.head())

# Afficher les premières lignes de l'ensemble de test
print(test_data.head())
```

```python
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, GRU

# Fonction pour calculer RMSE
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
##
# Forecast the product
def forecast_product1(product, order=(5,1,0)):
    # Extract training and testing data
    df_train = product_store[product]['train']
    df_test = product_store[product]['test']
    
    # Fit the ARIMA model
    model = ARIMA(df_train['Value'], order=order)
    model_fit = model.fit()
    
    # Forecast the test set
    forecast = model_fit.forecast(steps=len(df_test))
    
    # Get the actual value from the test set
    actual = df_test['Value'].values
    
    # Calculate the error
    error = rmse(actual, forecast.values)
    
    return forecast, actual, error
```

```python
# Test the function
results = {}
for product in product_store:
    print('Forecasting ', product)
    forecast, actual, error = forecast_product1(product)
    results[product] = {'forecast': forecast, 'actual': actual, 'error': error}
```

```python
# Calculate the overall error
overall_error = 0
for product in results:
    overall_error += results[product]['error']
overall_error /= len(results)
```

```python
for index, product in enumerate(results):
    print("Product: ", product)
    print("RMSE: ", results[product]['error'])
    print("---------------------------------------")
print('Overall RMSE: ', overall_error)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
def train_lstm_model(train_data, test_data):
    # Extraire les valeurs des colonnes 'value' des DataFrames
    train_values = train_data['Value'].values.reshape(-1, 1, 1)
    train_targets = train_data['Value'].values.reshape(-1, 1, 1)
    test_values = test_data['Value'].values.reshape(-1, 1, 1)
    
    # Définir le modèle
    model = Sequential([
        LSTM(32, input_shape=(1, 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    
    # Entraîner le modèle
    model.fit(train_values, train_targets, epochs=5, verbose=0)
    
    # Faire des prédictions sur les données de test
    predictions = model.predict(test_values)
    
    return predictions.flatten()
```

```python
# Entraîner et prédire pour chaque produit avec le modèle LSTM
lstm_results = {}
for product in product_store:
    print('Training LSTM for', product)
    train_data = product_store[product]['train']
    test_data = product_store[product]['test']
    predictions = train_lstm_model(train_data, test_data)
    
    # Extraire les valeurs de la colonne 'value' des DataFrames
    test_values = test_data['Value'].values
    # Calculer l'erreur RMSE
    error = rmse(test_values, predictions)
    
    lstm_results[product] = {'forecast': predictions, 'actual': test_values, 'error': error}
```

```python
# Calculate the overall error
overall_lstm_error = 0
for product in lstm_results:
    overall_lstm_error += lstm_results[product]['error']
overall_lstm_error /= len(lstm_results)
```

```python
for index, product in enumerate(lstm_results):
    print("Product: ", product)
    print("RMSE: ", lstm_results[product]['error'])
    print("---------------------------------------")
print('Overall RMSE: ', overall_lstm_error)
```

```python
from prophet import Prophet # type: ignore

# Fonction pour entraîner le modèle Prophet
def train_prophet_model(train_data, test_data):
    train_data = train_data.rename(columns={'Year': 'ds', 'Value': 'y'})
    test_data = test_data.rename(columns={'Year': 'ds', 'Value': 'y'})
    
    model = Prophet()
    model.fit(train_data)
    
    future = model.make_future_dataframe(periods=len(test_data), freq='Y')
    forecast = model.predict(future)
    
    forecast_values = forecast[-len(test_data):]['yhat'].values
    actual = test_data['y'].values
    error = rmse(actual, forecast_values)
    
    return forecast_values, actual, error
```

```python
# Entraîner et prédire pour chaque produit avec le modèle Prophet
prophet_results = {}
for product in product_store:
    print('Training Prophet for', product)
    train_data = product_store[product]['train']
    test_data = product_store[product]['test']
    forecast, actual, error = train_prophet_model(train_data, test_data)
    prophet_results[product] = {'forecast': forecast, 'actual': actual, 'error': error}

overall_prophet_error = 0
for product in prophet_results:
    overall_prophet_error += prophet_results[product]['error']
overall_prophet_error /= len(prophet_results)
print('Overall RMSE for Prophet: ', overall_prophet_error)
```
